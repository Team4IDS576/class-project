```python
import pandas as pd
links = pd.read_csv("NguyenLinks.csv")
routes = pd.read_csv("NguyenRoutes.csv")
flows = pd.read_csv("NguyenAssignment.csv")
```

```python
links["flow"] = flows["matrix_ab"]
```

```python
links.latency = links['free flow speed'] * (1 + (links['alpha'] * ((links['flow']/links['capacity']) ** links['beta'])))
```

```python
links
```

```python
# Function to calculate travel time for a route
def calculate_travel_time(route):
    link_nos = [int(link) for link in route.split(',')]
    travel_times = links[links['No'].isin(link_nos)]['latency']
    return travel_times.sum()

# Apply the function to calculate travel time for each route
routes['Travel Time'] = routes['Links sequence'].apply(calculate_travel_time)
```
